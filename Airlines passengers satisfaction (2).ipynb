```python
with sns.axes_style('white'):
    g = sns.catplot(x="Gate Location", y="Baggage Handling", hue="Satisfaction", col="Class", data=df, kind="bar", height=4.5)
```

For business class,it is observed that all gate locations have higher no. of dissatisfied passengers when baggage handling is not done perfectly well (rating <= 4).


For Eco Plus, when the gate location is 2, even though the baggage handling is done at mediocre level,some passengers still remain dissatisfied.
For Eco, when the gate location is 1, even though the baggages are handled properly with ratings range 3.0- 5.0 , passengers remained dissatisfied.

baggage handling service need to be improved in eco and eco plus to get ratings above 4.


```python
with sns.axes_style('dark'):
    g = sns.catplot(x="In-flight Wifi Service", y="In-flight Entertainment", hue="Satisfaction", col="Class", data=df, kind="box", height=4.5, aspect=.8)
```

```python
import matplotlib.pyplot as plt 
fig, axarr = plt.subplots(2, 2, figsize=(12, 8))

table1 = pd.crosstab(df['Satisfaction'], df['Check-in Service'])
sns.heatmap(table1, cmap='Oranges', ax = axarr[0][0])
table2 = pd.crosstab(df['Satisfaction'], df['In-flight Service'])
sns.heatmap(table2, cmap='Blues', ax = axarr[0][1])
table3 = pd.crosstab(df['Satisfaction'], df['On-board Service'])
sns.heatmap(table3, cmap='rainbow', ax = axarr[1][0])
table4 = pd.crosstab(df['Satisfaction'], df['Leg Room Service'])
sns.heatmap(table4, cmap='twilight', ax = axarr[1][1])
```

# For onboarding service,in-flight service, only 4 and 5 rating givers belong to satisfied passengers category.
